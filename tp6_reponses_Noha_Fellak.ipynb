# b)

On tranforme les valeurs de significativite avec uen trasnformation de log 10 pour plusieurs raisons. Par exemple, ces p-values etant tres petites, cela rends difficile leur visualisation et interpretation. La transformation log 10 permet de diminuer l'echelle et compacter ces valeurs dans un graphique. 

De plus, dans un echelle logarithmique, la difference entre des petites valeurs (comme 0.01 et 0.001) est plus prononcee que le difference entre de plus grandes valeurs (comme 1 et 0.1). Donc la transformation log 10 permet de mieux discerner entre des echantillons avec de hautes significances (donc des p-values tres petites).

On prends les TPM seulement des genes interagissants avec SARS. On trouve la mediane des TPM dans chaque tissu (donc le TPM le plus grand pour chaque tissu)

On trouve le z-score. on regarde la distribution de l'expression des proteines interagissant avec sars, sur la moyenne. Donc on regarde jusqu'a quel point l'expression de ces proteines est differente de la moyenne de toutes les proteines humaines.

On trouve la p-value (significativité de ce score Z) pour chaque tissu, et on le mets dans le graphique qu'on voit. 

# b)

On peut voir d'apres le graphe, que le Z-score des expressions des proteins interagissant avec sars, sur les expressions de toutes les proteines humaines est plus significatif pour des tissus plus que d'autres.

La p-value est la plus petite dans les tissus de poumons, donc pour les poumons, l'enrichissement en proteines avec lesquelles SARS interagit est le plus significatif. 

Cela fait du sens car la SARS-COV2 est une maladie principalement respiratoire, qui infecte a partir des voies aeriennes (donc ca fait du sens que les proteines avec lesquels elle interagit sont le plus exprimes dans les poumons)

Les resultats ici ne sont pas significatifs, car les p-values sont assez hausses.

A noter que puisque la transformation est un logarithme negatif, donc les autres valeurs du graphe inferieures a 0.6 vont en fait donner des p-values initiales encore plus grandes que 0.25.
